```python
import tifffile as tiff
import numpy as np
import glob
import os
import pandas as pd
import re

# Base directory containing subfolders 2021, 2022, 2023, 2024
base_folder = r"D:\Notes\7th Sem\Final Year Project\Raster\TIFF"
years = ["2021", "2022", "2023", "2024"]

data = []

for year in years:
    folder = os.path.join(base_folder, year)
    files = glob.glob(os.path.join(folder, "*.tif"))
    for f in files:
        img = tiff.imread(f).astype(np.float32)
        cloud_ratio = np.mean(img) * 100  # convert to percentage
        
        # Extract month-year info from filename
        name = os.path.basename(f)
        match = re.search(r"(\d{4})_(\d{2})", name)
        if match:
            month_year = f"{match.group(2)}-{match.group(1)}"
        else:
            month_year = "Unknown"

        data.append((name, month_year, cloud_ratio))

# Create DataFrame with serial numbers
df = pd.DataFrame(data, columns=["File_Name", "Month_Year", "Cloud_Cover(%)"])
df.index += 1
df.reset_index(inplace=True)
df.rename(columns={"index": "Sl_No"}, inplace=True)

# Save CSV
csv_path = os.path.join(base_folder, "cloud_cover_ratios.csv")
df.to_csv(csv_path, index=False)

# Print summary
for _, row in df.iterrows():
    print(f"{row['File_Name']} -> {row['Cloud_Cover(%)']:.2f}%")

print(f"\nDone. Results saved to: {csv_path}")
```
